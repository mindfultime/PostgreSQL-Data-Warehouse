```python
import psycopg2, pandas as pd
import matplotlib.pyplot as plt
from sql_queries import *

conn = psycopg2.connect("host=127.0.0.1 dbname=sparkifydb user=student password=student")
cur = conn.cursor()
```

```python
def select(query,conn):
    try:
        df=pd.read_sql_query(query, conn)
        conn.commit()
        return df
    except psycopg2.Error as e:
        print(e)
```

# User info by state
The graph represents user in the system by state

```python
df=select(user_location,conn)
df.plot(kind='bar',y='count',x='state',figsize=(20,5))
```

# Average Song Duration Per Artist Info
The graph represents avg song duration per artist in the system

```python
df=select(avg_song_duration_per_artist,conn)
df.plot(kind='bar',y='avg_duration',x='artist_name',figsize=(20,5))
```

# User Subscription Info
The graph represents the type of subscription (free/paid) of user in the system

```python
df=select(users_subscription,conn)
df.plot(kind='bar',x='subscription_type',figsize=(10,5))
```
